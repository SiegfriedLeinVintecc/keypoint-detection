```python
import torch.nn as nn
import torch
```

```python
m = nn.Sigmoid()
loss = nn.BCELoss()
input = torch.rand(1, 3, 256, 256, dtype=torch.float32)
target = torch.rand(1, 3, 256, 256, dtype=torch.float32)
output = loss(m(input), target)
print(input)
print(target)
print(output)
output = loss(target, target)
print(output)
sm = m(input)
output = loss(sm, sm)
print(output)
```
